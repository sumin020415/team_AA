```python
import pandas as pd
file_path = "./ref/광주.csv"
data = pd.read_csv(file_path, encoding = 'EUC-KR')
ss = data.iloc[:, 4:23].astype(int)
total = ss['05_06'] + ss['06_07'] + ss['07_08']+ ss['08_09']+ ss['09_10']+ ss['10_11']+ ss['11_12']+ ss['12_13']+ ss['13_14']+ ss['14_15']+ ss['15_16']+ ss['16_17']+ ss['17_18']+ ss['18_19']+ ss['19_20']+ ss['20_21']+ ss['21_22']+ ss['22_23']+ ss['23_24']
data['total'] = total
data
#ss['total'] = total
#print(ss.dtypes)
```

```python
columns = ['date', 'num', 'name', 'type']
time_col = [f'{i}~{i + 1}H' for i in range(5, 24)]
columns = columns + time_col + ['total']
data.columns = columns
```

```python
df = data.groupby([ 'date', 'num', 'name']).sum().reset_index()
df
```

```python
gdata = df[['date', 'name', 'total']]
gdata = gdata.sort_values(by = ['name', 'date'], ascending = [True, False])
gdata.head(10)
```

```python
mdata = gdata.groupby(['name']).sum().reset_index()
tdata = mdata[['name','total']]
tdata
```

```python
sdata = mdata.sort_values(by = 'total', ascending = False)
sdata.head(10)
total_data = sdata[:20]
total_data
```
